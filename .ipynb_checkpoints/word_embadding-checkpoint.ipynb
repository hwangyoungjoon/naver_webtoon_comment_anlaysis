```python
def mecab_NN_parse(text):
    parse_filter = re.compile(r'\S{2,8}\tNN')
    prep_text = parse_filter.findall(t.parse(text))
    nouns_text = [x.replace('\tNN', '') for x in prep_text]
    return nouns_text
```

```python
w2v_model.wv.syn0.shape   #(단어 1925개, 100차원) 
```

```python
from sklearn.preprocessing import normalize
word_vectors=normalize(w2v_model.wv.syn0,axis=1, norm="l2")
```

```python
kmeans= KMeans(n_clusters=6, max_iter=10, verbose=1)
print("kmeans.fit_predict()")
clustering=kmeans.fit_predict(word_vectors)
```

```python
from collections import defaultdict

cluster_to_row=defaultdict(lambda:[])
for row_id,label in enumerate(clustering):
    cluster_to_row[label].append(row_id)
    
cluster_to_row=dict(cluster_to_row)
for label,row in cluster_to_row.items():
    print("cluster # %d has %d comments" %(label,len(row)))
```

```python
count=defaultdict(lambda:0)
```

```python
docs=["파이콘에서 발표를 합니다.","대체 이 발표는 무엇인가?","나는 무엇을 발표해야하는가?"]
```

```python
for doc in docs:
    for word in doc.split():
        n=len(word)
        
        for e in range(1,n+1):
            count[word[:e]]+=1
            print(word[:e])
            print(count[word[:e]])
```

```python
word="아이오아이는"
n=len(word)
for e in range(2,n+1):
    w=word[:e]
    f=count[w]
    p=f/count[:e-1]
    
    print("{:6}, f={}, p={:.2}".format(w,f,p))
```
